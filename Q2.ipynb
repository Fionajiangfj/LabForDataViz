```python
import pandas as pd
df = pd.read_csv('mtcars.csv')
df.head()
```

## Question 1 (5 marks) Create a subset of the variables mpg, disp, hp and wt from the mtcars data. Construct the scatter plot for these attributes.

```python
import matplotlib.pyplot as plt

subset = df[['mpg', 'disp', 'hp', 'wt']]

pd.plotting.scatter_matrix(subset, alpha=0.8, figsize=(10, 10), diagonal='kde')
plt.show()
```

## Question 2 (10 marks) Fit a multiple linear regression model with mpg as the response variable and disp and hp as the input variables.

```python
import statsmodels.formula.api as sm

model = sm.ols('mpg ~ disp + hp', data=df).fit()

print(model.summary())
```

## Question 3 (5 marks) Check the p-values to evaluate the significance of the input variables.

```python
p_values = model.pvalues

print(p_values.round(3))
```

```python
# disp: p-value = 0.000, less than 0.05, significant.

# hp: p-value = 0.074, greater than 0.05, not significant.
```

## Question 4 (5 marks) Which independent variable is the least significant? Why?

**disp**: p-value = 0.000, less than 0.05, we reject the null hypothesis, so disp is a statistically significant predictor of mpg.

**hp**: p-value = 0.074, greater than 0.05, we do not have enought evidence to reject the null hypothesis, so hp is not a statistically significant predictor of mpg.

Thus, **hp** is the least significant among the two input variables.
