# Q1: Labels on 1-gallon cans of paint usually indicate the drying time and the area that can be covered in one coat. Most brands of paint indicate that, in one coat, a gallon will cover between 250 and 500 square feet, depending on the texture of the surface to be painted. One manufacturer, however, claims that a gallon of its paint will cover 400 square feet of surface area. To test this claim, a random sample of ten 1-gallon cans of white paint were used to paint 10 identical areas using the same kind of equipment. The actual areas (in square feet) covered by these 10 gallons of paint are given here:
#                               310, 311, 412, 368, 447, 376, 303, 410, 365, 350
# Do the data present sufficient evidence to indicate that the average coverage differs from 400 square feet?


# A: What are H0 and Ha ?

```python
# H0: µ = 400 (The mean coverage for this paint is 400 square feet.)
# HA: µ ≠ 400 (The mean coverage for this paint is not 400 square feet.)
```

# B: Find the p-value for the test, and use it to evaluate the statistical significance of the results.

```python
from scipy import stats

data = [310, 311, 412, 368, 447, 376, 303, 410, 365, 350]

t_statistic, p_value = stats.ttest_1samp(a= data, popmean=400)

print("T-statistic:", t_statistic)
print("P-value:", p_value)

'''
Interpret the results.

The two hypotheses for this particular one sample t-test are as follows:
Because the p-value of our test (0.0491) is less than alpha = 0.05,
we reject the null hypothesis of the test. We have sufficient
evidence to say that coverage for this paint is not 400 square feet.
'''
```

# Q2: Smoking and Lung Capacity In a study of the effect of cigarette smoking on the carbon monoxide diffusing capacity (DL) of the lung, researchers found that current smokers had DL readings significantly lower than those of either ex smokers or nonsmokers.
# The carbon monoxide diffusing capacities for a random sample of n = 20 current smokers are listed here:
# 103.768, 88.602, 73.003, 123.086, 91.052, 92.295, 61.675, 90.677, 84.023, 76.014 100.615, 88.017, 71.210, 82.115, 89.222, 102.754, 108.579, 73.154, 106.755, 90.479
# Do these data indicate that the mean DL reading for current smokers is significantly lower than 100 DL, the average for nonsmokers?


```python
data = [103.768, 88.602, 73.003, 123.086, 91.052, 92.295, 61.675, 90.677, 84.023, 76.014, 100.615, 88.017, 71.210, 82.115, 89.222, 102.754, 108.579, 73.154, 106.755, 90.479]

t_statistic, p_value = stats.ttest_1samp(a=data, popmean=100, alternative="less")

print("T-statistic:", t_statistic)
print("P-value:", p_value)

'''
Interpret the results.

The two hypotheses for this particular one sample t-test are as follows:
Because the p-value of our test (0.0033) is less than alpha = 0.05,
we reject the null hypothesis of the test. We have sufficient
evidence to say that the mean DL reading for current smokers is significantly lower than 100 DL.
'''
```

# Q2 :A Telecom service provider claims that individual customers pay on an average 400 rs. per month with standard deviation of 25 rs. A random sample of 50 customers bills during a given month is taken with a mean of 250 and standard deviation of 15. What to say with respect to the claim made by the service provider?

```python
from statsmodels.stats.weightstats import ztest

population_mean = 400
population_std = 25
sample_size = 50
sample_mean = 250
sample_std = 15

z_statistic, p_value = ztest(x1= [sample_mean] * sample_size, value=population_mean)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)

'''
Interpret the results.

Because the p-value of our test (close to 0) is less than alpha = 0.05,
we reject the null hypothesis of the test. We have sufficient
evidence to say that the mean bill for customers is not 400 rs.
The service provider's claim is incorrect.
'''
```

# Q7 Analyze the distribution of the mpg attribute and identify any outliers. If outliers are detected, replace them with the lower or upper quartile values accordingly.

```python
import numpy as np

Q1 = data['mpg'].quantile(0.25)
Q3 = data['mpg'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['mpg'] < lower_bound) | (data['mpg'] > upper_bound)]

if not outliers.empty:
  data.loc[(data['mpg'] < lower_bound), 'mpg'] = lower_bound
  data.loc[(data['mpg'] > upper_bound), 'mpg'] = upper_bound

print(data)
```
