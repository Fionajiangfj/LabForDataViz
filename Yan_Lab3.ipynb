Q1: A coin is biased so that the probability of heads is 2/3. What is the probability that exactly four heads come up when the coin is flipped seven times, assuming that the flips are independent?

```python
import scipy.stats as stats
from scipy.stats import binom

p = 2/3

n = 7

k = 4

print(binom.pmf(k, n, p))
```

Q3: A factory produces tools of which 98% are in good working order. Samples of 1000 tools are selected at random and tested.  
a) Find the mean and give it a practical interpretation.

b) Find the standard deviation of the number of tools in good working order in these samples.

```python
p = 0.98
n = 1000

mean = n * p
print("Mean:", mean)

# Practical interpretation: 980 tools out of 1000 samples are in good working order on average.

mean= binom.stats(n, p, moments = "m")
Variance = binom.stats(n, p, moments = "v")
skew = binom.stats(n, p, moments = "s")
kurt = binom.stats(n, p, moments = "k")

print("mean : ", mean)
print("variance : ", Variance)
print("skew : ", skew)
print("kurt : ", kurt)
```

Q4: Find the probability that at least 5 heads show up when a fair coin is tossed 7 times.

```python
p = 0.5
n = 7

print(binom.pmf(5, n, p) + binom.pmf(6, n, p) + binom.pmf(7, n, p))
```

Q5: A multiple choice test has 20 questions. Each question has four possible answers with one correct answer per question. What is the probability that a student will answer 10 or more questions correct (to pass) by guessing randomly?

```python
p = 1/4
n = 20

probability_at_least_10_correct = 0
for k in range(10, 21):
  probability_k_correct = binom.pmf(k, n, p)
  probability_at_least_10_correct += probability_k_correct

print(probability_at_least_10_correct)
```

Q6: A box contains 3 red balls, 4 white balls and 3 black balls. 6 times, a ball is selected at random, the color noted and then replaced in the box. What is the probability that the red color shows at least twice?

```python
n = 6
p = 3 / (3 + 4 + 3)

probability_at_least_twice = 0
for k in range(2, 7):
  probability_k_red = binom.pmf(k, n, p)
  probability_at_least_twice += probability_k_red

print(probability_at_least_twice)
```
