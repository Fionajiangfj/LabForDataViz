## 4. Analyze the monthly total profit and represent it with a histogram to highlight the most frequent profit ranges.

```python
plt.hist(df['total_profit'], bins=10, edgecolor='k')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Total Profit Frequency')
plt.show()
```

## 5. Determine the total sales for each product from last year and present the results in a pie chart.

```python
facecream = df['facecream'].sum()
facewash = df['facewash'].sum()
toothpaste = df['toothpaste'].sum()
bathingsoap = df['bathingsoap'].sum()
shampoo = df['shampoo'].sum()
moisturizer = df['moisturizer'].sum()

dataset = [facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer]
labels = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

plt.pie(dataset, labels = labels, autopct='%1.1f%%')
plt.legend(bbox_to_anchor = (1.6,1), loc='upper right')
plt.show()
```

## 6. Retrieve the monthly data for bathing soap and facewash, and display it using a subplot.

```python
month_number = df['month_number']
bathingsoap = df['bathingsoap']
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(month_number, bathingsoap)
plt.xlabel('Month')
plt.ylabel('Bathing Soap')

facewash = df['facewash']
plt.subplot(1,2,2)
plt.plot(month_number, facewash)
plt.xlabel('Month')
plt.ylabel('Facewash')

plt.show()
```
